```python
#hangman
#words to guess by user
#so far
#letters used
#1.chosw a word to be guessed
#2.we will create a loop
#3.ask user about his guess
#4.check if guesss in leter_used
#5.append the guess in leter use 
#6.check if their guess is right if so update the word so far
#7.if wrong print hangman updated
#8.if he have made 1 less mistake than the total characters of the words,game will end

hangman=(
    """
    -----
    |   |
    |
    |
    |
    |
    |
    |
----------
    """,
    """
    -----
    |   |
    |   O
    |  -+-
    |
    |
    |
    |
---------
    """,
    """
    ------
    |    |
    |    O
    |  /-+-/
    |   
    |
    |
    |
    |
----------
    """,
    """
    ------
    |    |
    |    O
    |  /-+-/
    |    |
    |    |
    |
    |
    |
    """,
    """
    -------
    |     |
    |     O
    |   /-+-/
    |     |
    |     |
    |    | |
    |    | |
    |
-----------
    """)
import random


words=['APPLE','ORACLE','LEMON','TESLA']
chosenwords=random.choice(words)
so_far='-'*len(chosenwords)

letter_used=[]
wrong_tries=len(chosenwords)-1
tries=0

while tries<wrong_tries and so_far!=chosenwords:
    print(hangman[tries])
    print('words so far: ',so_far)
    print("letters used: ",letter_used)
    guess=input('guess a letter: ').upper()
    
    while guess in letter_used:
        print('try again. you have already used this letter')
        guess=input('guess a letter: ').upper()
        
    else:
        letter_used.append(guess)
        if guess in chosenwords:
            print('you are doing great')
            new=''
            for i in range (len(chosenwords)):
                if guess==chosenwords[i]:
                    new=new+guess
                else:
                    new=new+so_far[i]
            so_far=new           
        else:
            print('incorrect letter! try again')
            tries=tries+1;
        
        
if tries==wrong_tries:
    print('unlucky, better luck next time')
else:
    print('congrats you win')
    
    
```
